```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x_axis=["Jan","Feb","Mar","April","May","June","July","Aug","Sep","Oct","Nov","Dec"]
y_axis=[10.02,23.24,39.20,35.42,32.24,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
 plt.plot(x_axis,y_axis)
```

```python
fig,ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
ax=plt.axes()
ax.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("Pyber Fare by Month")
plt.grid()
plt.legend()
```

```python
plt.bar(x_axis,y_axis)
```

```python
plt.bar(x_axis,y_axis,color="green",label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis)
```

```python
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
fig,ax=plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(x_axis,y_axis)
plt.scatter(x_axis,y_axis,color="red",label="Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
plt.pie(y_axis,labels=x_axis)     
plt.show()
```

```python
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode_values,labels=x_axis,autopct='%.1f%%')
```

```python
colors=["Slateblue", "magenta","lightblue","green","yellowgreen","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(6,6))
plt.pie(y_axis,
        explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```
